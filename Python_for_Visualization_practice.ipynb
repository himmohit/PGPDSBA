# * Histogram:- 
* we want to show histogram using current dataset.

```python
# Here we are using 'Histogram' Visualization in Price column. So we are first of all mentioning where data is and then what is the column want to histogram which is price.

sns.histplot(data =dt, x= 'price');
```

```python
# If we want to customize this histogram we can do by below mentioned point.

plt.title('Car Details')
plt.xlabel('Price of Car')
plt.xlim(3000,50000)
plt.ylabel('Total Cars')
plt.ylim(0,70)

sns.histplot(data=dt, x='price', color='orange');
```

```python
# Now we can change the number of bins (bars) there in the histogram by using below code:- 


plt.title('Car Details')
plt.xlabel('Price of Car')
plt.ylabel('Total Cars')

sns.histplot(data=dt, x='price', color='brown',bins=7);
```

```python
# Now we are using 'binwidth' to specify the width of bins or bars there.

#sns.histplot(data=dt, x='price', binswidth=30)

plt.title('Car Details')
plt.xlabel('Price of Car')
plt.ylabel('Total Cars')

sns.histplot(data=dt, x='price', color='brown',binwidth=20);
```

```python
# Again we change the width of bins to 800.

sns.histplot(data =dt, x= 'price', binwidth=800);
```

## - KDE (Kernel Density Estimation) in Histogram:- 
- It's line which get available just above the bars and as per their momentum it get change.

```python
from seaborn.distributions import KDE
# So here we are declairing 'KDE' in this histogram to add on a line on above bins to see how it works.

plt.title('Car Details')
plt.xlabel('Price of Car')
plt.ylabel('Total Cars')

sns.histplot(data=dt, x='price', color='green', kde=True);
```

```python
# Extracting first 5 rows data from Height of Car Column.

dt['height'].head()
```

```python
# In this we are creating histogram in Car height column.

plt.title('Height of Car')
plt.xlabel('Height of Car in inches')
plt.ylabel('Total Cars')

sns.histplot(data=dt, x='height', color='cyan', bins=18,kde=True);
```

## - Using of 'Hue' in Histogram:- 
* With the help of 'Hue' we can add another column in single bar and then it'll show value as per available data.

```python
# We are using 'Hue' in histogram while using 'hue' we are adding another 'fuel_type' column to see the changes.

plt.title('Height of Car')
plt.xlabel('Price of Car with Fuel Details|')
plt.ylabel('Total Cars')

sns.histplot(data=dt, x='price', hue='fuel_type', color='yellow',kde=True);
```

```python
# Again we are applying 'Hue' in 'engine_type' column.

plt.title('Height of Car')
plt.xlabel('Height of Car in inches')
plt.ylabel('Total Cars')

sns.histplot(data=dt, x='height', hue='engine_type', color='cyan',kde=True);

# As we can see the visualization is little bit messy. So to avoid that we'll use another concept below 'FacetGrid'.
```

## - To avoid messy visualization in above histogram we are using this concept of 'FacetGrid': 
* In this it'll create multiple histogram as per the category available in another column.

```python
# So here we are using 'Facetgrid' Concept it'll create multiple histogram as per the available category in column 'fuel_type'.

fcg =sns.FacetGrid(dt, col='engine_type') 
fcg.map(sns.histplot,'price', color='grey',kde=True);
```

# * Boxplot:- 

* It's ploting chart available in Seaborn library it can visualize available data in Box Plot. Also it contain Minimum and Max Value along with 25% (Q1), 50%(Median), 75%(Q3) of data you can see below in picture. Apart from that it help us to find outliers in available columns. Outliers always left from Min and Right side of Max.

boxplot.png

```python
# Here we are using 'Boxplot' visualization from seaborn library to display the 'price' column and find outliers.

sns.boxplot(data=dt, x='price')
```

```python

plt.title('Boxplot for Normalized_losses ')


sns.boxplot(data=dt, x='normalized_losses', color='purple')
```

```python
dt[dt['horsepower'] == dt['horsepower'].max()]
```

## - Skewed Data View using 'Boxplot':-

* **Right Skewed** - In this Median is goes near to Q1 and it shows less distance between Min and Q1.
* **Left Skewed** - In this Median is more towards to Q3 and it shows less distance between Q3 and Max.

skew_box.png

```python
# Here we are displaying 'horsepower' column with the help of 'Boxplot' library. So you can see in this minimum horsepower available in car is 50, Max is somewhere around 180
# 25% of car having horsepower around 70, 50% of car having horsepower around 90 apart from that 75% of car have horsepower around 110. Even you can see there's outliers 
# available which is more than maximum value..

plt.title('Boxplot for "Horsepower"')
plt.xlabel('Horsepower')

sns.boxplot(data=dt, x='horsepower', color='red')

# So here in this it's 'Right-Skewed' Data because it's 'Median' part is near to 'Q1' and also have shortest distance between 'Min' and 'Q1'.
```

# * Bar Graph:- 

* It use to display value in categorical column with the help of bins.
* Formulae for 'Bar Graph' is **'countplot()'**.

```python
# Here it's showing Total how many Car available with which 'engine_type'.

sns.countplot(data=dt, x='engine_type')
```

```python
# Here it's showing Total how many Car available as per the engine placed 'engine_location'.

sns.countplot(data=dt, x='engine_location')
```

```python
# So now we are compairing Car 'body style' with 'number of cylinder'.

plt.figure(figsize=(30,7)) 
sns.countplot(data=dt, x='body_style', hue='number_of_cylinders');
```

```python
# Here it's showing Total Car available as per 'manufacturer'.

plt.title('Total Car as per Manufacturer')
plt.xlabel('Total Car')
plt.ylabel('Manufacturer')

sns.countplot(data=dt, x='make');
# It's seems messy here so below lines it's try to make figure size bigure and also rotate 'xticks' to 90 degree.
```

```python
# Here it's showing Total Car available as per 'manufacturer'.

plt.figure(figsize=(15,7));
plt.title('Total Car as per Manufacturer')

sns.countplot(data=dt, x='make');

# As you can see it's showing little bit messy in 'Manufacturer'. So we'll rotate that into 90 degree.
```

```python
# Here it's nice and clear with information.

plt.figure(figsize=(15,7));
plt.title('Total Car as per Manufacturer')
plt.xticks(rotation=90)
sns.countplot(data=dt, x='make');
```
